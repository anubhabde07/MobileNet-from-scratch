```python
from tensorflow.keras.preprocessing.image import load_img, img_to_array
import numpy as np
import matplotlib.pyplot as plt
from keras.models import load_model
```

```python
# Load the model
model = load_model('model_plant_classification.keras')
```

```python
def detect_and_display(image_path, model):
    try:
        # Load and preprocess image
        img = load_img(image_path, target_size=(224, 224))
        img_array = img_to_array(img)
        img_array = np.expand_dims(img_array, axis=0)
        img_array = img_array / 255.0

        # Prediction
        pred = model.predict(img_array)
        pred_class = int(np.argmax(pred))
        confidence = float(pred[0][pred_class]) * 100

        class_names = [
            "aloevera",
            "banana",
            "bilimbi",
            "cantaloupe",
            "cassava",
            "coconut",
            "corn",
            "cucumber",
            "curcuma",
            "eggplant",
            "galangal",
            "ginger",
            "guava",
            "kale",
            "longbeans",
            "mango",
            "melon",
            "orange",
            "paddy",
            "papaya",
            "peperchili",
            "pineapple",
            "pomelo",
            "shallot",
            "soybeans",
            "spinach",
            "sweetpotatoes",
            "tobacco",
            "waterapple",
            "watermelon"
        ]

        # Safety check
        if pred_class >= len(class_names):
            raise ValueError("Prediction index out of class range")

        result = class_names[pred_class]

        # Display
        plt.figure(figsize=(5, 5))
        plt.imshow(img)
        plt.title(f"Predicted Class: {result} ({confidence:.2f}%)")
        plt.axis("off")
        plt.show()

    except Exception as e:
        print("Error:", e)
```

```python
# example
image_path = r"C:\Users\deanu\Downloads\Orange_img.webp"
detect_and_display(image_path, model)
```
